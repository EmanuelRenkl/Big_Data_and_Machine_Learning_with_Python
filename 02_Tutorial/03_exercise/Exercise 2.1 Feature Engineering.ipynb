```python
# Load the Boston House Price dataset using sklearn.datasets.load_boston
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.keys())
print(boston["feature_names"])
len(boston["feature_names"])
```

```python
import pandas as pd
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df.head()
```

```python
# Extract polynomial features (without bias) and interactions up to a degree of 2.
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree =2, include_bias=False)
```

```python
values = df.values.reshape(-13,13)
values
```

```python
print(poly.fit_transform(values))
```

```python
print(poly.get_feature_names())
```

```python
# How many features do you end up with?
len(poly.get_feature_names())
```

```python
# Create a pandas DataFrame using the polynomials.
df2 = pd.DataFrame(poly.fit_transform(values))
print(df2)
```
